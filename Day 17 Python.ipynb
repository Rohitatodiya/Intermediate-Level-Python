```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image

fname = r'your.png'

#opening image using pil 
image = Image.open(fname).convert('L')
#mapping image to gray scale
plt.title("Data Analysis")
plt.xlabel("length")
plt.ylabel("Width")
plt.imshow(image, cmap = 'Spectral')
plt.show()
```

# Matplotlib Pie

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["english","hindi","maths","science"]
c = ['yellow','black','blue','magenta']
plt.pie(x, labels = y, colors = c)
plt.show()
```

```python
x = [10,20,30,40]
y = ["english","hindi","maths","science"]
c = ['yellow','black','blue','magenta']
plt.pie(x, labels = y, colors = c)
plt.legend()
plt.show()
```
