# Linear Graph

```python
x = [1,2,3,4]
y = [2,4,5,6]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,5,6]
c = 'g'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,6,7]
y = [10,5,8,4,2]
c = 'y'
plt.plot(x,y,c, marker = 'o')
plt.show()
```

```python
x = [5,2,9,6,7]
y = [10,5,8,4,2]
c = 'y'
plt.plot(x,y)
plt.plot(y, marker = '>')
plt.plot(x, marker = 'o')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']

plt
plt.bar(x,y, color = c)
plt.show()
```

```python
import numpy as np
```

```python

x = np.array(["A","B","C","D"])
y = np.array([5,6,7,8])
c = ['r','g','b','y']

plt.barh(x,y, color = c)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([5,6,7,8])


plt.barh(x,y, color = "Pink")
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([5,6,7,8])


plt.bar(x,y, color = "black", width =0.5)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([5,6,7,8])


plt.barh(x,y, color = "black", height =0.3)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1,2,3,4,5,6]
y = [ 2,4,5,6,8,1]
plt.title("Scatter plot")

plt.scatter(x,y)
plt.show()
```

```python
x = [ 1,2,3,4,5,6]
y = [ 2,4,5,6,8,1]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")

plt.scatter(x,y)
plt.show()
```

```python
x = [ 1,2,3,4,5,6]
y = [ 2,4,5,6,8,1]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
c = ['red','magenta','green','orange','blue','brown']

plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [ 1,2,3,4,5,6]
y = [ 2,4,5,6,8,1]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
c = ['red','magenta','green','orange','blue','brown']

plt.scatter(x,y, color = c, s = 200)
plt.show()
```
