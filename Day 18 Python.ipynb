# Seaborn

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,4,3,2,1]
sns.lineplot(x= var1, y =var2)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"var1":var1,"var2":var2})
sns.lineplot(x = 'var1',y = 'var2', data =df)
plt.show()
```
